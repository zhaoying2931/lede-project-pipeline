```python
def liquid_map(df, mapbox_token, show=True):
    liquid = go.Scattermapbox(
                              lat = df['LOCATION_LATITUDE'],
                              lon = df['LOCATION_LONGITUDE'],
                              mode = 'markers',
                              text = df['INCIDENT_IDENTIFIED_DATETIME'],
                              name = 'COMMODITY_RELEASED_TYPE',
                              hovertemplate='%{text} <extra></extra>',
                              below = '',
                              marker=dict(size = 10, color = '#BA3A0B', opacity = 0.9)
    )
    
    layout= go.Layout(
                       width = 600,
                       height = 600,
                       showlegend = True,
                       legend = dict(font = dict(
                                                family = 'Roboto',
                                                size =10, color = '#114B5F'),
                                    orientation = 'h',
                                    yanchor='top', x = 0.025
                                    ),
                       paper_bgcolor = '#F8FAFA',
                       plot_bgcolor = '#F8FAFA',
                       hoverlabel = dict(font_size = 10, font_family = 'Roboto'),
                       hovermode = 'closest',
                       margin = dict(l = 10, r = 10, t = 70, b = 10),
                       mapbox = dict(center = dict(lat = 39.27500, lon = 76.18667),
                                    accesstoken = mapbox_token,
                                    zoom = 4.8,
                                    style = 'light')
    )
    
    fig = go.Figure(data = df,layout = layout)   
    #setting configurations
    if show:
        config = dict(
            {
                #I don't want Plotly's toolbar to be displayed
                "displayModeBar": False,
                "scrollZoom": True,
                "displaylogo": False,
                "responsive": True,
                "staticPlot": False,
            }
        )
        
        fig.show(config=config)
```

```python
liquid_map(map_df, df_token, show = True)
```
