```python
import plotly.express as px
import pandas as pd
```

```python
df = pd.read_csv('data/df_liquid.csv')
df.info()
```

```python
px.set_mapbox_access_token(open('mapbox_token.ipynb').read())
```

```python
fig = px.scatter_mapbox(df, 
                        lat = 'LOCATION_LATITUDE', 
                        lon = 'LOCATION_LONGITUDE', 
                        color = 'Commodity_Released_Type', 
                        size = 'net_spill',
                        opacity = 0.6,
                        hover_name = 'Operator',
                        hover_data = ['Cause','Estimated Property Damage Cost'],
                        color_continuous_scale = px.colors.cyclical.IceFire, 
                        zoom = 10)
fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0},  
        hoverlabel=dict( 
        bgcolor="white",     # white background
        font_size=12,        # label font size
        font_family="Inter") # label font
)
    
fig.show()
```
